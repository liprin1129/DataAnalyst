Because there are 24 samples and we do not know the mean and standard deviation of population, t-statistics will be used to identify this hypothesis test. It is assumed the samples follows the central limit theorem, so the distributions are Gaussian. Also, this test will be an independent test, because the samples are not involved with repeated measures, such as longitudinal, and pre- and post-test. Both samples are aquired from arbitrary people, and they do not have influence to each other.

### 4. Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

People tend to take more time to finish the task on the incongruent words than that on the congruent words. In addition, it became wider for the standard deviation of incongruent task than that of congruent task.
Though the above table tells some descriptive statistics, it is hard to identify whether the hypothesis is significant or not, as well as which kind of central tendency has more appropriate descriptive statistics for each dataset.

Therefore, it is better to plot a histogram for each dataset so as to find the most descriptive one.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("HW1_dataset.csv")

data = data.drop('Unnamed: 0', axis=1)

fig, ax = plt.subplots(2, 1, figsize = [6, 8])
x_range = np.ones(100)

#####################
## Congruent words ##
#####################

## Histogram
c_min = np.round(np.min(data['Congruent'])) - 2
c_max = np.round(np.max(data['Congruent'])) + 2
c_range = np.arange(c_min, c_max, 2.3)
# print(c_range)

ax[0].hist(data['Congruent'], bins = c_range, color='Cyan', alpha=0.3, label='Subjects')
ax[0].set_title('Congruent')
ax[0].set_xlim(5, 25)
ax[0].set_ylim(0, 8)

## Mean, Median, Mode
c_mean = 14.04
c_std = 3.56
c_median = 14.36

y_mean_range = np.linspace(0, 6.6, 100)
ax[0].plot(x_range*c_mean, y_mean_range, '-', c='g', linewidth = 2, label="Mean (14.04)")
y_median_range = np.linspace(0, 6.55, 100)
ax[0].plot(x_range*c_median, y_median_range, '-', c='r', linewidth = 2, label="Median (14.36)")

## Distribution
c_tickRange = np.linspace(5, 25, 100)
c_pdf = stats.norm.pdf(c_tickRange, c_mean, c_std) * 59 # now get theoretical values in our interval
ax[0].plot(c_tickRange, c_pdf, c='black', label="Distribution", linewidth = 2)

## Labels
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Count')
ax[0].legend(loc='best', fontsize=10)

#######################
## Incongruent words ##
#######################

## Histogram
i_min = np.round(np.min(data['Incongruent'])) - 2
i_max = np.round(np.max(data['Incongruent'])) + 2
i_range = np.arange(i_min, i_max, 2.3)
# print(i_range)

ax[1].hist(data['Incongruent'], bins = i_range, color='Cyan', alpha=0.3, label='Subjects')
ax[1].set_title('Incongruent')
ax[1].set_xlim(12, 37)
ax[1].set_ylim(0, 8)

## Mean, Median, Mode
i_mean = 22.02
i_median = 21.02
y_mean_range = np.linspace(0, 4.9, 100)
ax[1].plot(x_range*i_mean, y_mean_range, '-', c='g', linewidth = 2, label='Mean (22.02)')
y_median_range = np.linspace(0, 5.6, 100)
ax[1].plot(x_range*i_median, y_median_range, '-', c='r', linewidth = 2, label='Median (21.02)')

## Distribution
i_tickRange = np.linspace(15, 37, 200)
ag,bg,cg = stats.gamma.fit(data['Incongruent'])
pdf_gamma = stats.gamma.pdf(i_tickRange, ag, bg,cg) * 59
ax[1].plot(i_tickRange, pdf_gamma, c='black', label="Distribution", linewidth = 2)

## Labels
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Count')
ax[1].legend(loc='best', fontsize=10)

fig.tight_layout()
```

The histogram of the congruent words dataset has 7 bins with 2.3 bin size, and it looks like a normal distribution. Theoretically, the mean, median, and mode of its sample dataset should be the same value. 

However, though incongruent words dataset has the same bin size, 2.3, as the congruent histogram with 9 bins, it seems a positive skewed distribution with one outlier bin and it has an influence the incongruent mean value. The green line (mean value) is slightly on right to the red line (median value). Therefore, IQR must be checked by boxplot to be convinced with the most appropriate kind of central tendency.

```python
fig, ax = plt.subplots(1, 1, figsize = [7, 5])

x_range = np.linspace(1, 2, 100)

i_IQR = ax.boxplot(data['Incongruent'])
i_boxes = i_IQR['boxes'][0].get_data()[1]

y_Q1 = np.ones(100) * 18.71675
ax.plot(x_range, y_Q1, '-.', linewidth=0.5, c='black')

y_Q2 = np.ones(100) * 24.0515
ax.plot(x_range, y_Q2, '-.', linewidth=0.5, c='black')

y_Min = np.ones(100) * 15.687
ax.plot(x_range, y_Min, '-.', linewidth=0.5, c='black')

y_Max = np.ones(100) * 26.282
ax.plot(x_range, y_Max, '-.', linewidth=0.5, c='black')

ax.text(1.34, 15.9, 'Min (15.69)')
ax.text(1.35, 19, 'Q1 (18.72)')
ax.text(0.73, 20.7, 'Median (21.02)', color='r')
ax.text(1.35, 24.3, 'Q2 (24.05)')
ax.text(1.33, 26.5, 'Max (26.28)')
ax.set_xticks([])
ax.set_title('Box Plot for Incongruent Dataset')
```

Then, it is obvious that the incongruent words dataset has a positive skewed distribution with the 1st quartile (18.72), 2nd quartile (24.05), Min (15.69), Max (26.28), and two outliers, 35.255 and 34.288 respectively.

* The mean difference, $\bar{\mathbf{d}}$, is $7.96$.
* The standard deviation of the differences, $s_d=\frac{\sum \mathbf{d}}{\sqrt{n-1}}=\frac{7.96}{\sqrt{24-1}}$, is $1.66$.
* Standard error of the mean difference, $SE(d)=\frac{s_d}{\sqrt{n}}=\frac{1.66}{\sqrt{24}}$, is $0.34$, which means the sample means is differ from the true population mean by 0.34.
* Because the degree of freedom $n-1$ is 23, $t_{critical}$ in positive directional one-tailed having at $\alpha=0.05$ is $1.714$.
* $t_{statistics}=\frac{\bar{\mathbf{d}}}{SE(d)}$ is $23.41$, this is in $t_{critical}$ region.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline

mean = 7.98
std = 0.79

t_critical = np.ones(100) * 1.714
t_statistics = np.ones(100) * 23.41
y = np.linspace(0, 0.6, 100)

plt.plot(t_critical, y, c='r', linewidth=2)
plt.plot(t_statistics, y, c='b', linewidth=2)

# Distribution
tickRange = np.linspace(-10, 30, 1000)
c_pdf = stats.norm.pdf(tickRange, 0, std)
plt.plot(tickRange, c_pdf, c='black', label="Distribution", linewidth = 2)

plt.title('Result')
plt.text(2.1, 0.5, 't critical (1.714)', color='red')
plt.text(12, 0.3, 't statistics (23.41)', color='blue')
```
